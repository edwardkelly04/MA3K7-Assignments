```python
#Rough Approximation
import random

def game_sim(n):
    result = 0
    for _ in range(n):
        result += random.choice([1, 2])
    return result

#Test
for i in range(10):
  n = 16
  final_result = game_sim(n)
  print(f"final result: {final_result}")
```

```python
#100,000 testing
def random_tester(n):
    result = 0
    for _ in range(n):
        result += random.choice([1, 2])
        if result == n:
            return True
    return False

counter = 0
for i in range(100000):
    if random_tester(24):
        counter += 1

print(f"24 reached {counter} times.")
print(f"Probability: {counter/100000}")
```

```python
#A million test
counter = 0
for i in range(1000000):
    if random_tester(24):
        counter += 1

print(f"24 reached {counter} times.")
print(f"Probability: {counter/1000000}")
```

```python
#Manual P(E) calculation
import math

P_E = 0


for n in range(0,13):
  P_E += (math.factorial(24-n))/((math.factorial(24-(2*n)))*math.factorial(n)) * 0.5**(24-n)

print(P_E)
```

```python
#Review

#Random Specialisation

n = random.randint(1, 48)
counter = 0

for i in range(100000):
    if random_tester(n):
        counter += 1

print(f"{n} was reached {counter} times.")
```

```python
#P(n) Plot

import numpy as np
import matplotlib.pyplot as plt

freq = np.zeros(49)

for n in range(1,49):
  for i in range(100000):
      if random_tester(n):
          freq[n] += 1

prob = freq/100000
final_prob = prob[1:]
x_vals = range(1,49)
plt.scatter(x_vals,final_prob)
plt.xlabel("n")
plt.ylabel("P(n)")
plt.title("Probability density function for landing on n")
plt.plot(x_vals,[2/3]*48)
plt.legend(["Probability approximation","P(n)=2/3"])
plt.show()
```

```python
#Testing for '3 game'

def tri_tester(n):
    result = 0
    for _ in range(n):
        result += random.choice([1, 2, 3])
        if result == n:
            return True
    return False

n = random.randint(1, 72)
counter = 0

for i in range(100000):
    if tri_tester(n):
        counter += 1

print(f"{n} was reached {counter} times.")
```

```python
#P(n) plot for '3 game'
freq = np.zeros(73)

for n in range(1,73):
  for i in range(100000):
      if tri_tester(n):
          freq[n] += 1

prob = freq/100000
final_prob = prob[1:]
x_vals = range(1,73)
plt.scatter(x_vals,final_prob)
plt.xlabel("n")
plt.ylabel("P(n)")
plt.title("Probability density function for landing on n")
plt.plot(x_vals,[0.5]*72)
plt.legend(["Probability approximation","P(n)=0.5"])
plt.show()
```

```python
#'x game' P(24) test

def n_tester(n):
    result = 0
    for _ in range(24):
        result += random.choice(range(1,n+1))
        if result == 24:
            return True
    return False

freq = np.zeros(24)


for n in range(1,24):
  for i in range(100000):
      if n_tester(n):
          freq[n] += 1

prob = freq/100000
final_prob = prob[1:]
x_vals = range(1,24)
plt.scatter(x_vals,final_prob)
plt.plot(x_vals,(1/(np.log(x_vals)))-0.25)
plt.xlabel("x")
plt.ylabel("P(24)")
plt.title("Probability of landing on 24 for each 'x game'")
plt.legend(["P(24) approximation","P(24)= 1/ln(x) - 0.25 "])
plt.show()
```
