```python
import numpy as np
import matplotlib.pyplot as plt


#Entry - Plot 1
def x(n):
    return 1/4 * (1-((1/3)**n))


n_vals = np.arange(0, 101)
f_vals = x(n_vals)


plt.plot(n_vals, f_vals, label=r'$x_{n}$')
y_vals = np.full(n_vals.shape, 0.25)
plt.plot(n_vals, y_vals, label='y = 0.25', linestyle=':', color='red',markersize=1, marker='o')


plt.title('Volume of non-empty glass n (3-game)')
plt.xlabel('n')
plt.ylabel(r'$x_{n}$')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#Plot 3
#Plot of V_F + Plot with estimate

def V_F(m):
    n_vals = np.arange(1, m // 2)
    sum = np.sum(1 / (2 * n_vals))
    V_F = 0.5 + sum
    return V_F


m_vals = np.arange(4, 101, 2)
V_F_vals = [V_F(m) for m in m_vals]

plt.plot(m_vals, V_F_vals, marker='o', label=r'$\mathbf{V}_F$')
plt.title(r'Plot of $\mathbf{V}_F$ over m')
plt.xlabel(r'$m$')
plt.ylabel(r'$\mathbf{V}_F$')
plt.grid(True)
plt.legend()
plt.show()


estimate = 0.35*(np.log(m_vals - 3)) + 1

plt.plot(m_vals, V_F_vals, label=r'$\mathbf{V}_F$', marker='o')
plt.plot(m_vals, estimate, label=r'$0.35\ln(m - 3) + 1 $', marker='x')

plt.title(r'Plot of $\mathbf{V}_F$ over m')
plt.xlabel(r'$m$')
plt.ylabel(r'Value')

plt.grid(True)
plt.legend()
plt.show()
```

```python
import random

#Table Values - Extension 1

#Game Simulation
def game(n):
    size = n // 2
    arr = np.full(size, 0.5)
    while len(arr) > 1:
        arr = arr[:-1]
        arr += 0.5 / len(arr)
    return arr[0]

#Estimate
def est(n):
  return 0.35*(np.log(n - 3)) + 1


random_vals = random.sample(range(4, 101, 2), 10)

# Print results for each random m value
for m in random_vals:
    result = game(m)
    estimate = est(m)
    diff = abs(result - estimate)

    print(f"For m = {m}:")
    print(result)
    print(estimate)
    print(diff)
    print()
```

```python
#Checks for Conj1, 2

#Extension 1 Check for initial check

#Conjecture 1 check
for m in m_vals:
  a = game(m)
  b = game(m+1)
  if a != b:
    print(m)
print('done')

#Conjecture 2 check
for m in m_vals:
  a = game(m)
  b = game(m+1)
  if a <= 1:
    print(m)
  if b<= 1:
    print(m+1)
print('done')
```

```python
#Extension 2 plots

def V_F1(m):
    n_vals = np.arange(1, m // 2)
    sum = np.sum(1 / (4 * n_vals))
    V_F = 0.25 + sum
    return V_F

def V_F2(m):
    n_vals = np.arange(1, m // 2)
    sum = np.sum(1 / (2 * n_vals))
    V_F = 0.5 + sum
    return V_F

def V_F3(m):
    n_vals = np.arange(1, m // 2)
    sum = np.sum(3 / (4 * n_vals))
    V_F = 0.75 + sum
    return V_F

def V_F4(m):
    n_vals = np.arange(1, m // 2)
    sum = np.sum(1 / n_vals)
    V_F = 1 + sum
    return V_F

m_vals = np.arange(4, 101, 2)

V_F1_vals = [V_F1(m) for m in m_vals]
V_F2_vals = [V_F2(m) for m in m_vals]
V_F3_vals = [V_F3(m) for m in m_vals]
V_F4_vals = [V_F4(m) for m in m_vals]

plt.plot(m_vals, V_F1_vals, label=r'$\bf{X}=0.25$',marker='o')
plt.plot(m_vals, V_F2_vals, label=r'$\bf{X}=0.5$',marker='o')
plt.plot(m_vals, V_F3_vals, label=r'$\bf{X}=0.75$',marker='o')
plt.plot(m_vals, V_F4_vals, label=r'$\bf{X}=1$',marker='o')

plt.title(r'Plot of $\mathbf{V}_F$ over m for $\bf{X}=0.25,0.5,0.75,1$')
plt.xlabel(r'$m$')
plt.ylabel(r'Value')

plt.grid(True)
plt.legend()
plt.show()

est1 = 0.5*(0.35*(np.log(m_vals - 3)) + 1)
est3 = 1.5*(0.35*(np.log(m_vals - 3)) + 1)
est4 =2*(0.35*(np.log(m_vals - 3)) + 1)


#Plot 1
plt.plot(m_vals, V_F1_vals, label=r'$\bf{X}=0.25$', marker='o')
plt.plot(m_vals, est1, label=r'$\frac{f(m)}{2}$', marker='x')
plt.title(r'Plot of $\mathbf{V}_F$ over m for $\bf{X}=0.25$')
plt.xlabel(r'$m$')
plt.ylabel(r'Value')

plt.grid(True)
plt.legend()
plt.show()

#Plot 2
plt.plot(m_vals, V_F3_vals, label=r'$\bf{X}=0.75$', marker='o')
plt.plot(m_vals, est3, label=r'$\frac{3f(m)}{2}$', marker='x')
plt.title(r'Plot of $\mathbf{V}_F$ over m for $\bf{X}=0.75$')
plt.xlabel(r'$m$')
plt.ylabel(r'Value')

plt.grid(True)
plt.legend()
plt.show()

#Plot 3
plt.plot(m_vals, V_F4_vals, label=r'$\bf{X}=1$', marker='o')
plt.plot(m_vals, est4, label=r'2f(m)', marker='x')
plt.title(r'Plot of $\mathbf{V}_F$ over m for $\bf{X}=1$')
plt.xlabel(r'$m$')
plt.ylabel(r'Value')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#Extension 3 plots

def Lb(m):
    n_vals = np.arange(1, m // 2)
    sum = np.sum(1 / n_vals)
    total = 1 + sum
    Lb = 1 / total
    return Lb

m_vals = np.arange(4, 101, 2)
Lb_vals = [Lb(m) for m in m_vals]

#Initial Plot
plt.plot(m_vals, Lb_vals, marker='o')
plt.title(r'Plot of $\mathbf{L}_B$ over m')
plt.xlabel(r'$m$')
plt.ylabel(r'$\bf{L}_B$')

plt.grid(True)
plt.show()


est_Lb =1/(2*(0.35*(np.log(m_vals - 3)) + 1))


#Plot with guess
plt.plot(m_vals, Lb_vals, label=r'$\bf{L}_B$', marker='o')
plt.plot(m_vals, est_Lb, label=r'$\frac{1}{2f(m)}$', marker='x')
plt.title(r'Plot of $\mathbf{L}_B$ over m')
plt.xlabel(r'$m$')
plt.ylabel(r'$\bf{L}_B$')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#Checks for conj 4 and 5

#Game Simulation
def new_game(n):
    size = (n-2) // 4
    arr = np.full((2*size)+1, 0.25)
    while len(arr) > 1:
        arr = arr[:-2]
        arr += 0.5 / len(arr)
    return arr[0]


#Known cases testing
print(new_game(6))
print(new_game(10))
print(new_game(14))

new_m_vals = np.arange(6, 101, 4)

#Conjecture 1 check
for m in new_m_vals:
  a = new_game(m)
  b = new_game(m+1)
  c = new_game(m+2)
  d = new_game(m+3)
  if not (a == b == c == d):
    print(m)
print('done')

#Conjecture 2 check
for m in new_m_vals:
  a = new_game(m)
  b = new_game(m+1)
  c = new_game(m+2)
  d = new_game(m+3)
  if a <= 1:
    print(m)
  if b<= 1:
    print(m+1)
  if c<= 1:
    print(m+2)
  if d<= 1:
    print(m+3)
print('done')

```
