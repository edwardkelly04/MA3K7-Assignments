```python
import numpy as np

#Returns a random nxn matrix with either equal 0's and 1's or one more 1
def ran_mat(n):
    total = n * n
    zeros = total // 2 + total % 2 
    ones = total - zeros
    elements = [0] * zeros + [1] * ones
    np.random.shuffle(elements)
    matrix = np.array(elements).reshape(n, n)
    return matrix

#Prints the winning matricies out of 10 random 
i_vals = range(10)
for i in i_vals:
    A = ran_mat(4)
    if np.linalg.det(A)==0:
        for row in A:
            print(" ".join(map(str, row)))
        print()
        



```

```python
import random

#Game Simulation
def game_P0(n):
    matrix = np.full((n, 2), -1)
    matrix[0][0] = 0 # Player 0's first go
    for i in range(n):
        player_1(matrix)
        check_win(matrix)
        player_0(matrix)
        
        
def game_P1(n):
    matrix = np.full((n, 2), -1)
    for i in range(n):
        player_1(matrix)
        player_0(matrix)
        check_win(matrix)

        
        
def player_1(matrix):
    #Random move out of availible moves
    moves = [(i, j) for i in range(len(matrix)) for j in range(len(matrix[0])) if matrix[i][j] == -1]
    if moves:
        row, col = random.choice(moves)
        matrix[row][col] = 1
        

def player_0(matrix):
    placed = False 
    
    #Checking to see if 0 can be placed adjacent to previous 1 
    for i, row in enumerate(matrix):
        if 1 in row:  
            for j, k in enumerate(row):
                if k == -1:  
                    matrix[i][j] = 0
                    placed = True
                    break  
        if placed:
            break 
            
    #If not then 0 placed randomly
    if not placed: 
        moves = [(i, j) for i in range(len(matrix)) for j in range(len(matrix[0])) if matrix[i][j] == -1]
        if moves:
            row, col = random.choice(moves)
            matrix[row][col] = 0
            

def check_win(matrix):
    if np.any(matrix == -1):
        return  # Do nothing
    
    column_sums = matrix.sum(axis=1)
    if np.all(column_sums == 1):
        print("Player 0 wins")
    else:
        print("Player 1 wins")

#Run the game for each player starting for 4<=n<=100
for n in range(4,101):
    game_P0(n)
    game_P1(n)

```
